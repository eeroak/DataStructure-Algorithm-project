```python
from ipynb.fs.full.BinarySearch import BinarySearch
from ipynb.fs.full.LinearSearch import LinearSearch
from ipynb.fs.full.SentinelLinearSearch import sentinelLinearSearch
import matplotlib.pyplot as plt
import timeit as ti
import random
```

## All times are represented in seconds!

#### Testing speeds of:
Binary search (O(log n), O(1))  
Linear search (O(n), O(1))  
Sentinel linear search (O(n), O(1))  
-algorithms  

###### Note that some of the calculations are so fast that it counts the seconds as e^-x

```python
# Generating data for the search, all the search functions will be tested on this data
# The data has to be the same for all the search functions to ensure that the results are comparable
random.seed(33) # Setting seed to ensure that the data is the same every time the code is run
data1k = [random.randint(0, 1000) for i in range(1000)]
data10k = [random.randint(0, 10000) for i in range(10000)]
data100k = [random.randint(0, 100000) for i in range(100000)]
data1M = [random.randint(0, 1000000) for i in range(1000000)]
x1k = random.randint(0, 1000) # Random number to search for in the data
x10k = random.randint(0, 10000) 
x100k = random.randint(0, 100000)
x1M = random.randint(0, 1000000)

print("Data generated successfully","\n",x1k,"\n",x10k,"\n",x100k,"\n",x1M)
```

```python
# Linear search
timer_start = ti.default_timer()
LinearSearch(data1k, x1k)
timer_end = ti.default_timer()
print("Linear search on 1k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
LinearSearch(data10k, x10k)
timer_end = ti.default_timer()
print("Linear search on 10k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
LinearSearch(data100k, x100k)
timer_end = ti.default_timer()
print("Linear search on 100k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
LinearSearch(data1M, x1M)
timer_end = ti.default_timer()
print("Linear search on 1M data took: ", timer_end - timer_start)
```

```python
# Binary search
timer_start = ti.default_timer()
BinarySearch(data1k, 0, len(data1k)-1, x1k)
timer_end = ti.default_timer()
print("Binary search on 1k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
BinarySearch(data10k, 0, len(data10k)-1, x10k)
timer_end = ti.default_timer()
print("Binary search on 10k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
BinarySearch(data100k, 0, len(data100k)-1, x100k)
timer_end = ti.default_timer()
print("Binary search on 100k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
BinarySearch(data1M, 0, len(data1M)-1, x1M)
timer_end = ti.default_timer()
print("Binary search on 1M data took: ", timer_end - timer_start)
```

```python
# Sentinel Linear search
timer_start = ti.default_timer()
sentinelLinearSearch(data1k, len(data1k), x1k)
timer_end = ti.default_timer()
print("Sentinel Linear search on 1k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
sentinelLinearSearch(data10k, len(data10k), x10k)
timer_end = ti.default_timer()
print("Sentinel Linear search on 10k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
sentinelLinearSearch(data100k, len(data100k), x100k)
timer_end = ti.default_timer()
print("Sentinel Linear search on 100k data took: ", timer_end - timer_start)

timer_start = ti.default_timer()
sentinelLinearSearch(data1M, len(data1M), x1M)
timer_end = ti.default_timer()
print("Sentinel Linear search on 1M data took: ", timer_end - timer_start)
```

```python
# Plotting the results
plt.style.use('fivethirtyeight')

x = ['1k', '10k', '100k', '1M']
linear = [0.0000411, 0.0000484, 0.0019, 0.0048]
binary = [0.000030, 0.0000225, 0.0000235, 0.0000222]
sentinel = [0.00005, 0.00006, 0.002, 0.005]
plt.plot(x, linear, label = 'Linear Search', linewidth=2, linestyle='-', color='blue')
plt.plot(x, binary, label = 'Binary Search', linewidth=2, linestyle='--', color='red')
plt.plot(x, sentinel, label = 'Sentinel Linear Search', linewidth=2, linestyle='-.', color='green')
plt.xlabel('Data size (number of elements)', fontsize=12)
plt.ylabel('Time taken (seconds)', fontsize=12)
plt.title('Search algorithm speed comparison', fontsize=14)

plt.legend()
plt.show()
```
