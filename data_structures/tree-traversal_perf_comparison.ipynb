```python
from ipynb.fs.full.tree_traversal import in_order_traversal, pre_order_traversal, post_order_traversal
from ipynb.fs.full.trees import Node
import seaborn as sns
import timeit as ti
import random
```

## All times are represented in seconds!

#### Testing speeds of:
In-order  
Pre-order  
Post-order  
-traversal algorithms  

##### They all have time and space complexities of O(n), and O(h), dependant on the amount of nodes and height of the tree

```python
# Generating a tree for testing with random 100 nodes
root = Node(50)
for i in range(1, 100):
    root.insertData(random.randint(1, 100))
root.visualize()
```

```python
# Testing the time taken for in-order traversal
start = ti.default_timer()
in_order_traversal(root)
end = ti.default_timer()
result_in = end - start
print(f"Time taken for in-order traversal: {end - start}")
```

```python
# Testing the time taken for pre-order traversal
start2 = ti.default_timer()
pre_order_traversal(root)
end2 = ti.default_timer()
result_pre = end2 - start2
print(f"Time taken for pre-order traversal: {end - start}")
```

```python
# Testing the time taken for post-order traversal
start3 = ti.default_timer()
post_order_traversal(root)
end3 = ti.default_timer()
result_post = end3 - start3
print(f"Time taken for post-order traversal: {end - start}")
```

```python
# Plotting the results
sns.barplot(x=["In-Order", "Pre-Order", "Post-Order"], y=[result_in, result_pre, result_post]).set(title="Time taken for tree traversal with 100 nodes", xlabel="Traversal type", ylabel="Time taken (s)")
sns.set_style("whitegrid")
sns.set_palette("bright")
sns.set_context("notebook")
sns.despine()
sns.set(rc={"figure.figsize":(10, 5)})
sns.set(rc={"axes.facecolor":"#f5f5f5"})
sns.set(rc={"axes.grid":True})
```
